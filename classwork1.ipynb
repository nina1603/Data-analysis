```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import glob
```

```python
df_cards = []
for filename in glob.glob('trainingData_JSON_chunk1.json'):
    with open(filename) as fin:
        for record in map(json.loads, fin):
            for card in record['player'].get('hand', []):
                card['turn'] = record['turn']
                card['gamestate_id'] = record['gamestate_id']
                card['hero'] = record['player']['hero']['hero_card_id']
                df_cards.append(card)
df = pd.DataFrame.from_records(df_cards)
```

```python
df_stat = []
for filename in glob.glob('trainingData_JSON_chunk1.json'):
    with open(filename) as fin:
        for record in map(json.loads, fin):
            d = {}
            d['player'] = record['player']['hero']['hero_card_id']
            d['opponent'] = record['opponent']['hero']['hero_card_id']
            d['result'] = record['decision']
            df_stat.append(d)
df = pd.DataFrame(df_stat)
```
